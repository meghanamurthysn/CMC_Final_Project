```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
import json
import datetime
import os
import matplotlib.dates as mdates
from matplotlib.dates import WeekdayLocator
from matplotlib.dates import MO
import datetime
```

```python
x = list()
y = list()
for root, dirs, files in os.walk("C:\\Users\\rke dell lati\\Documents\\MIMS\\Fall 2016\\Info 216\\Final Project\\206"):
    for file in files:
        if file.endswith(".json"):
            with open(os.path.join(root, file),"r") as jsondata:
                data = json.load(jsondata)
                json_normalize(data)
                count = 0
                for each in data:
                    count += 1
                dt = datetime.datetime.fromtimestamp(float(each["ts"])).date()
                x.append(dt)
                y.append(count)
xs = list()
ys = list()
def next_weekday(d, weekday):
    days_ahead = weekday - d.weekday()
    if days_ahead <= 0: # Target day already happened this week
        days_ahead += 7
    return d + datetime.timedelta(days_ahead)
end_dt = datetime.date(2016, 11, 19)
d = datetime.date(2016, 9, 1)
while d + datetime.timedelta(7)<=end_dt:
    next_tuesday = next_weekday(d, 1) # 0 = Monday, 1=Tuesday, 2=Wednesday...
    xs.append(next_tuesday)
    d += datetime.timedelta(7)
    if next_tuesday in x:
        idx = x.index(next_tuesday)
        ys.append(y[idx])
    else:
        ys.append(0)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
# plot
ax.plot(x,y,lw=5,color='orange')
ax.scatter(xs,ys,s=360,c='red',alpha=0.8)
# beautify the x-labels
plt.gcf().autofmt_xdate()
plt.ylabel('Messages', fontsize=18)
ax.xaxis.set_major_locator(WeekdayLocator(byweekday=MO))
labels = ax.get_xticklabels()
labels1 = ax.get_yticklabels()
plt.setp(labels, rotation=30, fontsize=12)
plt.setp(labels1, rotation=0, fontsize=12)
ax.fill_between(x, 0, y, facecolor='orange', alpha=0.25)
ax.grid(True)
title_obj = plt.title('Slack Channel for course usage')
plt.getp(title_obj)                    
plt.getp(title_obj, 'text')
plt.setp(title_obj, color='g', fontsize=18)

ax.spines['bottom'].set_color('green')
ax.spines['top'].set_color('green')
ax.spines['left'].set_color('green')
ax.spines['right'].set_color('green')
ax.xaxis.label.set_color('green')
ax.tick_params(axis='x', colors='green')
ax.yaxis.label.set_color('green')
ax.tick_params(axis='y', colors='green')
plt.show()
```
