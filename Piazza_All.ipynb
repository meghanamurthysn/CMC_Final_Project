```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import math
```

```python
df = pd.read_csv('piazza-info206_stats-Dec-06.csv', low_memory=False)
```

```python
instr_df = df.loc[df['role'] == 'Instructor']
instr_contri = instr_df.contributions
instr_names = instr_df.name

fig, ax = plt.subplots(figsize=(2, 2))
x_pos = np.arange(0, 1.5*len(instr_names), 1.5)
performance = instr_contri
width = 0.5

ax.bar(x_pos, performance, width, align='center', color='orange')
ax.set_xticks(x_pos)
ax.set_xticklabels(instr_names)
plt.ylabel('Contributions', fontsize=8)
title_obj = plt.title('Contributions by Instructors')
plt.getp(title_obj)                    
plt.getp(title_obj, 'text')
plt.setp(title_obj, color='g', fontsize=8)

ax.spines['bottom'].set_color('green')
ax.spines['top'].set_color('green')
ax.spines['left'].set_color('green')
ax.spines['right'].set_color('green')
ax.xaxis.label.set_color('green')
ax.tick_params(axis='x', colors='green')
ax.yaxis.label.set_color('green')
ax.tick_params(axis='y', colors='green')

plt.show()
```

```python
student_df = df.loc[df['role'] == 'Student']
student_contri = student_df.contributions
student_names = student_df.name

fig, ax = plt.subplots(figsize=(10, 5))
x_pos = np.arange(0, 1.5*len(student_names), 1.5)
performance = student_contri
width = 1.5

ax.bar(x_pos, performance, width, align='center', color='orange')
ax.set_xticks(x_pos)
ax.set_xticklabels(student_names, fontsize = 8, rotation = 45)
plt.ylabel('Contributions', fontsize=8)
title_obj = plt.title('Contributions by Students')
plt.getp(title_obj)                    
plt.getp(title_obj, 'text')
plt.setp(title_obj, color='g', fontsize=8)

ax.spines['bottom'].set_color('green')
ax.spines['top'].set_color('green')
ax.spines['left'].set_color('green')
ax.spines['right'].set_color('green')
ax.xaxis.label.set_color('green')
ax.tick_params(axis='x', colors='green')
ax.yaxis.label.set_color('green')
ax.tick_params(axis='y', colors='green')

plt.show()
```

```python
student_df = df.loc[df['role'] == 'Student']
student_answers = student_df.answers
student_names = student_df.name

fig, ax = plt.subplots(figsize=(10, 5))
x_pos = np.arange(0, 1.5*len(student_names), 1.5)
performance = student_answers
width = 1.5
yint = range(min(performance), math.ceil(max(performance))+1)
ax.bar(x_pos, performance, width, align='center', color='orange')
ax.set_xticks(x_pos)
ax.set_xticklabels(student_names, fontsize = 8, rotation = 45)
ax.set_yticks(yint)
plt.ylabel('Answers')
title_obj = plt.title('Answers by Students')
plt.getp(title_obj)                    
plt.getp(title_obj, 'text')
plt.setp(title_obj, color='g', fontsize=8)

ax.spines['bottom'].set_color('green')
ax.spines['top'].set_color('green')
ax.spines['left'].set_color('green')
ax.spines['right'].set_color('green')
ax.xaxis.label.set_color('green')
ax.tick_params(axis='x', colors='green')
ax.yaxis.label.set_color('green')
ax.tick_params(axis='y', colors='green')

plt.show()
```
